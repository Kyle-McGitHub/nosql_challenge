### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
hyg_count = establishments.count_documents(query)
print("Number of documents with the desired hygiene score:", hyg_count)

# Display the first document in the results using pprint
hyg_document = establishments.find_one(query)
pprint(hyg_document)
```

```python
# Convert the result to a Pandas DataFrame

hyg_raw_df = pd.DataFrame(list(establishments.find(query)))

# Split the scores column into separate columns
scores_df = hyg_raw_df['scores'].apply(pd.Series)

# Concatenate the original DataFrame with the new scores DataFrame
hyg_df = pd.concat([hyg_raw_df, scores_df], axis=1)

hyg_df = hyg_df[['BusinessName', 'BusinessType', 'BusinessTypeID', 'Hygiene']]

# Display the number of rows in the DataFrame
print("Number of rows in the DataFrame:", len(hyg_df))

# Display the first 10 rows of the DataFrame
hyg_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "LocalAuthorityName": {"$regex": "London"},
    "RatingValue": {"$gte": 4}
}

# Use count_documents to display the number of documents in the result
result_count = establishments.count_documents(query)
print("Number of documents with London as Local Authority and RatingValue >= 4:", result_count)

# Display the first document in the results using pprint
rating_document = establishments.find_one(query)
pprint(rating_document)
```

```python
# Convert the result to a Pandas DataFrame
#df = pd.DataFrame(list(establishments.find(query)))
# Display the number of rows in the DataFrame
#print("Number of rows in the DataFrame:", len(df))
# Display the first 10 rows of the DataFrame
#df.head(10)

df = pd.DataFrame(list(establishments.find(query)))

rating_df = df[['BusinessName', 'BusinessType', 'BusinessTypeID', 'RatingValue']]

# Display the number of rows in the DataFrame
print("Number of rows in the DataFrame:", len(hyg_df))

# Display the first 10 rows of the DataFrame
rating_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

query_penang_flavours = {"BusinessName": "Penang Flavours"}
penang_flavours = establishments.find_one(query_penang_flavours)

degree_search = 0.01
latitude = penang_flavours['geocode']['latitude']
longitude = penang_flavours['geocode']['longitude']

query = {
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search},
    "RatingValue": 5
}

sort = [("Hygiene", 1)]
limit = 5

results = establishments.find(query).sort(sort).limit(limit)

result_dicts = []

for result in results:
    
    result_dicts.append(result)

print(result_dicts)
```

```python
# Convert result to Pandas DataFrame
top_5_raw_df = pd.DataFrame(result_dicts)

# Split the scores column into separate columns
sort_df = top_5_raw_df['scores'].apply(pd.Series)

# Concatenate the original DataFrame with the new scores DataFrame
top_5_df = pd.concat([top_5_raw_df, sort_df], axis=1)

top_5_df = top_5_df[['BusinessName', 'BusinessType', 'BusinessTypeID', 'RatingValue', 'Hygiene']]

top_5_df = top_5_df.sort_values(by='Hygiene', ascending=True)

top_5_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Print the first 10 results

pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",
            "count": {"$sum": 1}
        }
    },
    {
        "$sort": {
            "count": -1
        }
    }
]

results = list(establishments.aggregate(pipeline))

print("Number of documents in the result:", len(results))

for result in results[:10]:
    print(result)
```

```python
# Convert the result to a Pandas DataFrame
pipeline_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f"Number of rows: {len(pipeline_df)}")

# Display the first 10 rows of the DataFrame
pipeline_df.head(10)
```
